```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/cricket.csv')
df.head()
```

```python
df.shape
```

```python
#renaming the columns
df = df.rename(columns= {'Mat':'Matches','NO':'Not_out', 'HS':'Highest_score', 'Ave':'Avg','BF':'Balls_faced', 'SR':'Strike_rate', '100':'100s','50':'50s', '0':'Ducks'})
df
```

```python
#checking null values
df.isnull().any()
```

```python
df[df['Balls_faced'].isna()==1]
```

```python
#filling null values
df[df['Strike_rate'].isna()==1]
```

```python
df['Balls_faced'] = df['Balls_faced'].fillna(0)
df['Strike_rate'] = df['Strike_rate'].fillna(0)
```

```python
df[df['Player'] == 'ED Weekes (WI)']
```

```python
df[df['Player'] == 'Hon.FS Jackson (ENG)']
```

```python
df[df['Player'] == 'CL Walcott (WI)']
```

```python
#dropping duplicates
df = df.drop_duplicates()
```

```python
df['Span'].str.split(pat= '-')
```

```python
df[['Start_year', 'Final_year']] = df['Span'].str.split('-',2, expand= True)
df
```

```python
#dropping Span column
df.drop(['Span'], axis = 1, inplace= True)
```

```python
#splitting Player column
df['Player'].str.split(pat='(')
```

```python
# creating Country column
df['Country'] = df['Player'].str.split(pat='(').str[1]
df['Country'] = df['Country'].str.split(pat=')').str[0]
```

```python
df['Player'] = df['Player'].str.split(pat= '(').str[0]
df
```

```python
#clearing unwanted values in Highest_score column
df['Highest_score'].str.split(pat = '*').str[0]
```

```python
# changing datatype
df['Highest_score'] = df['Highest_score'].astype('int')
```

```python
# changing datatype
df['Start_year'].astype('int')
```

```python
#clearing unwanted values in Balls Faced column
df['Balls_faced'] = df['Balls_faced'].str.split(pat='+').str[0]
df
```

```python
# changing datatype
df['Balls_faced'] = df['Balls_faced'].astype(int)
```

```python
# changing datatype
df.astype({'Start_year': 'int', 'Final_year': 'int'})
```

```python
df.dtypes
```

```python
# creating new column Career Lenght
df['Career_length'] = df['Final_year'] - df['Start_year']
```
