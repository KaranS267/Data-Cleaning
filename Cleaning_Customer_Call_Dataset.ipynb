```python
import pandas as pd
df = pd.read_excel("/content/Customer Call List.xlsx")
df
```

```python
df.shape
```

```python
#cleaning name
df["Last_Name"] = df["Last_Name"].str.strip("123/*-_.")
df
```

```python
# dropping columns and duplicates
df = df.drop(columns= "Not_Useful_Column")
df = df.drop_duplicates()
df
```

```python
# cleaning phn number
df["Phone_Number"] = df["Phone_Number"].str.replace('[^0-9a-zA-Z]', '')
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# cleaning phn number
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', ' ')
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', ' ')
df
```

```python
# cleaning address
df[["Street_address", "State", "Pincode"]] = df["Address"].str.split(",", 2, expand=True)
df
```

```python
# cleaning/replacing the values
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
```

```python
# cleaning/replacing the values
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
# filling na values
df = df.fillna('')
df
```

```python
# dropping the do not contact numbers
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace=True)

df
```

```python
# dropping the na phn numbers

for x in df.index:
  if df.loc[x, 'Phone_Number'] == ' ':
    df.drop(x, inplace= True)

df
```

```python
df.reset_index(drop= True)
```

```python
df.to_csv("Cleaned_customer_call List.csv")
```
